```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/ANN/Churn_Modelling.csv')
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(X), print(y)
```

```python
dataset['Exited'].value_counts()
```

## Encoding The categorical data

```python
# Label Encoding Gender Label
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X[:,2] = le.fit_transform(X[:,2])
```

```python
print(X)# Female - 0 ad Male - 1
```

## One Hot Encoding 'geography' column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers= [('encoder',OneHotEncoder(),[1])],remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
print(X)
```

## Splitting the dataset into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

##Building The ANN
* Initializing the ANN
* Adding the input layer and first hidden layer
* Adding the second hidden layer
* Adding the output layer

```python
# Initializing the ANN
ann = tf.keras.models.Sequential()
```

```python
# Adding the input layer and first hidden layer    # Activation func - rectifier(relu)
ann.add(tf.keras.layers.Dense(units =6,activation = 'relu'))
```

```python
# Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units =6,activation = 'relu'))
```

```python
# Adding the output layer    # Activation func - sigmoid()
ann.add(tf.keras.layers.Dense(units=1,activation = 'sigmoid'))
```

## Training the ANN

# Compiling The ANN

```python
ann.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

# Training the ANN on training set

```python
ann.fit(X_train,y_train,batch_size = 32,epochs = 100)
```

## Making the Predictions and evaluating the model

```python
# Predicting the result of single observation
Use the ANN model to predict if the customer with the following info will leave the bank:
Geography : France
Credit Score : 600
Gender : Male
Age : 40 yrs old
Tenure : 3 Years
Balance : $ 60000
Number of Products : 2
Does this customer have a credit card? Yes
Is this customer an ActiveMember : Yes
Estimated Salary: $50000
So, should we say goodbye to that customer?
```

```python
# Note 1 : Notice that the values of the features were all the input in a doube pair of square brackets.
# that's because the 'predict' method always expects a 2D array as the format of its inputs. And putting our values
# into a double pair of square brackets makes the input exactly a 2D array

# Note 2 : Notice also that the France country was not input as string in the last column but as '1 0 0' n the first three
# Columns . Because of course the predict method expects the one hot encoded value of the state and we see in the first row
# Of the matrix of features X, "France" encoded as '1,0,0' and we need to take care of fummy variables
```

```python
# Make a prediction
prediction = ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]]))
print(prediction[0]*100)
```

```python
ann.evaluate(X_test,y_test)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred>0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)

```
